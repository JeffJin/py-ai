```python
new_data1 = {
    'name': ['cat', 'tiger', 'lion'],
    'age': [22, 33, 21],
    'height': [178, 99, 88]
}

new_data2 = {
    'name': ['chicken', 'fly', 'pig'],
    'age': [2, 3, 1],
    'height': [17, 9, 8]
}

df_one = pd.DataFrame(new_data1)
df_two = pd.DataFrame(new_data2)

df_combined = pd.concat([df_one, df_two], ignore_index=True)
# df_combined.reset_index(inplace=True)

row_data = {
    'name': 'mouse',
    'age': 2,
    'height': 78,
    'description': 'mouse description',
}

column_data = {
    'description': ['cat description', 'tiger description', 'lion description'],
}

df_three = pd.DataFrame(column_data)

df_combined = pd.concat([df_combined, df_three], axis=1)
df_combined.loc[len(df_combined)] = row_data

df_combined.to_csv('./data/animals-2.csv', index=None, header=False)
```
