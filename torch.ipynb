```python
# import random
import requests
import torch
x = torch.rand(5, 3)
# print(x, torch.__version__)
torch.set_default_dtype(torch.float64)

tensor_arr = torch.Tensor([[5, 3], [4, 3]]).type(torch.IntTensor)

print(torch.is_tensor(tensor_arr), tensor_arr, torch.numel(tensor_arr))

uninitialized = torch.Tensor(3, 5).type(torch.int64)
print(uninitialized)

# Define indices and values for a 3x3 sparse tensor
indices = torch.tensor([[0, 1, 2],
                        [1, 2, 2]], dtype=torch.int64) # Non-zero elements at (0,0), (1,1), (2,2)
values = torch.tensor([1.0, 2.0, 3.0])

# Construct the sparse COO tensor
sparse_tensor = torch.sparse_coo_tensor(indices, values, size=(3, 3))

print(sparse_tensor)
print(sparse_tensor.to_dense()) # Convert to dense for visualization
```
