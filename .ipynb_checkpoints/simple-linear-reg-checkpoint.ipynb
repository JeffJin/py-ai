```python
print("Hello World!")
#import libraray
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#loading the data set
sample_data =pd.read_csv('./data/salary_Data.csv')
sample_data.head()

sample_data.info()
```

```python
sample_data.describe()

sample_data.skew()

plt.figure(figsize=(12,6))
sns.histplot(x=sample_data['YearsExperience'], bins=20, kde=True, color='orange')

```

```python

plt.figure(figsize=(12,8))
sns.boxplot(y=sample_data['YearsExperience'], color='azure')
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(y=sample_data['Salary'], color='maroon')
```

```python
Q1 = sample_data.quantile(0.25, axis=0)
print('First_quantile :', Q1)
```

```python
Q3 = sample_data.quantile(0.75, axis=0)
print('Third_quantile :', Q3)
```

```python
IQR = Q3 - Q1
print('Inter Quartile Range :', IQR)
```
